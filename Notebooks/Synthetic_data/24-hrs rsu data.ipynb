```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

df_hubs = pd.read_csv(r"C:\Users\LIKITH REDDY\Applied data science\Project\stations_data.csv")

def generate_live_sim_data(hubs_df):
    all_logs = []
    start_time = datetime.now().replace(hour=0, minute=0, second=0, microsecond=0)
    
    for i in range(96): # 15-min intervals for a day => 24 * 4 = 96
        current_time = start_time + timedelta(minutes=15 * i)
        
        hour = current_time.hour
        occ_rate = 0.85 if (9 <= hour <= 11 or 18 <= hour <= 21) else 0.4
        
        for _, hub in hubs_df.iterrows():
            total = int(hub['charging_points'])
            occupied = np.random.binomial(total, occ_rate)
            
            all_logs.append({
                "timestamp": current_time,
                "station_id": hub['station_id'],
                "occupied_now": occupied,
                "free_now": total - occupied
            })
            
    return pd.DataFrame(all_logs)

live_sim = generate_live_sim_data(df_hubs)
live_sim.to_csv("rsu_live_feed.csv", index=False)
```
