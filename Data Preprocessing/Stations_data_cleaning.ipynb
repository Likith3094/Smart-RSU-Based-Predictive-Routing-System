```python
import pandas as pd
import numpy as np
import os 
print(os.getcwd())
stations_data = pd.read_csv(r"C:\Users\LIKITH REDDY\Applied data science\Project\raw_data\ev_stations.csv")
#rsu_data = pd.read_csv(r"C:\Users\LIKITH REDDY\Applied data science\Project\rsu_historical_data.csv")
```

```python
stations_data.isnull().sum()
needed_columns = ['latitude','longitude','open','close','city','total']
for col in stations_data.columns:
    if col not in needed_columns:
        del stations_data[col]
print(stations_data.columns)

stations_data.isna().sum()
stations_data['city'].value_counts()
print(stations_data.head())
```

```python
#adding new columns
stations_data = stations_data[stations_data['city'].isin(['Delhi', 'New Delhi'])]
stations_data.rename(columns={'total': 'charging_points'}, inplace=True) 
stations_data['cost_per_unit'] = np.random.randint(18, 24, size=len(stations_data))
print(stations_data['latitude'].value_counts())
stations_data['longitude'].value_counts()
```

```python
import pandas as pd

stations_data = stations_data[(stations_data['charging_points'] > 0) & (stations_data['latitude'] != 0) & (stations_data['longitude'] != 0)]

# 1. Group by coordinates to collapse all individual points into one station
stations_data = stations_data.groupby(['latitude', 'longitude']).agg({
    
    'charging_points': 'sum',  # Total ports at this exact location
}).reset_index()

# 2. Assign a clean, unique Station ID for your RSU system
stations_data['station_id'] = [f'STATION_{i+1:04d}' for i in range(len(stations_data))]

# 3. Reorder for a cleaner view
stations_data = stations_data[['station_id', 'latitude', 'longitude','charging_points']]

# Verify the new structure
print(stations_data.head())
print(f"\n dataset now has {len(stations_data)} unique stations.")
```

```python
import pandas as pd

# 1. Filter out rows with 0 charging points or 0 coordinates
stations_data = stations_data[(stations_data['charging_points'] > 0) & (stations_data['latitude'] != 0) & (stations_data['longitude'] != 0)]

# 2. Group by location to create unique Station Hubs
stations_data_clean= stations_data.groupby(['latitude', 'longitude']).agg({
    'charging_points': 'sum'
}).reset_index()

# 3. Assign final Station IDs
stations_data_clean['station_id'] = [f'STATION_{i+1:04d}' for i in range(len(stations_data_clean))]
stations_data_clean = stations_data_clean[['station_id', 'latitude', 'longitude', 'charging_points']]

print(f"Valid stations after cleaning: {len(stations_data_clean)}")
```
